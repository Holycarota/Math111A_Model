```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
data = pd.read_csv("output.csv")
data
```

```python
x = data.get("X")
y = data.get("Y")
v = data.get("Velocity")
Distance = data.get("Distance_to_goal")
IfLeft = data.get("IfLeft")
vertical = data.get("Vertical_Angle")
```

```python
#fig,ax = plt.subplots()
plt.plot(Distance,v,zorder=1)
plt.grid()
plt.xlabel("Distance")
plt.ylabel("Velocity")
plt.scatter(30,30,s=50, color="r",zorder=2)
plt.scatter(33,29,s=50, color="r",zorder=2)
plt.axvline(x=38, color = "r", zorder = 2)
plt.show()
```

As we can see in the graph, as the Distance to goal increases, the overall velocity needed also increases. Note that overall is increasing, but that doesn't necessary mean if d1>d2, then v2>v1. Also, the min velocity needed is increasing. For distance around 30 to 40, the velocity is varied, or we can say flexible.

```python
plt.scatter(x,y, c=v,s=50,cmap=cm.summer)
cbar=plt.colorbar()
cbar.set_label("Velocity")
plt.title("Max Velocity Needed with respect to X and Y",fontsize=15)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

For this graph, when the shooting point is getting closer and getting towards middle part of the field, the max velocity needed decreases.

```python
plt.scatter(x,y,c=IfLeft,cmap=cm.summer)
cbar=plt.colorbar()
cbar.set_label("1 is left, and 0 is right")
plt.title("Left Corner or Right Corner", fontsize=15)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

As we can see from the image, we can saparate the image into four parts. The first part would be -33<x<0 and 15<y<35, which is almost all color in Green. The second part is 0<x<33 and 15<y<35, which is almost all colors in yellow.

```python
#Calculate the third part
third_part = data[(data["X"]>=-33) & (data["X"]<0) & (data["Y"]>=0) & (data["Y"]<=15)]
third_part.get("IfLeft").mean()
```

For this area, 46% time the model will suggest targeting left top corner, which means, 54% time targeting right top corner is better. Since this is symmetrical, for the last part, 46% time the model will suggest aiming right top corner, and 56% time to left top corner

```python
plt.scatter(x,y,c=vertical,cmap=cm.summer)
cbar=plt.colorbar()
cbar.set_label("Vertical_Angle")
plt.title("Vertical Angle needed", fontsize=15)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

For this image, which is kind of interesting, when the distance is closed to goal, the vertical angle needed is high. When it is getting farer, the vertical angle is decreasing. However, at some points, it is getting higher again. (Need graph to explain about). The reason is not hard to understand. Since the trajectory of the ball is a hyperbola, in the 2d coordinates, there will be two x points, with resepect to one y point. So, when the vertical angle is increasing and the shooting location is really closed to the goal, we need the ball to pass the defensive wall and the goal line, these two points will be at the part where first derivative of the hyperbola is greater than 0, or the vertical displacement of the ball is still increasing. When the vertical angle needed hits the lowest requirement, which is around 14 degree, and it starts getting higher again. This time, the location of the goal line is at the part where the ball is falling, or we can say the first derivative is smaller than 0.
