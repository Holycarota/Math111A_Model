```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
data = pd.read_csv("output.csv")
data
```

```python
x = data.get("X")
y = data.get("Y")
v = data.get("Velocity")
Distance = data.get("Distance_to_goal")
```

```python
#fig,ax = plt.subplots()
plt.plot(Distance,v,zorder=1)
plt.grid()
plt.xlabel("Distance")
plt.ylabel("Velocity")
plt.scatter(30,30,s=50, color="r",zorder=2)
plt.scatter(33,29,s=50, color="r",zorder=2)
plt.axvline(x=38, color = "r", zorder = 2)
plt.show()
```

As we can see in the graph, as the Distance to goal increases, the overall velocity needed also increases. Note that overall is increasing, but that doesn't necessary mean if d1>d2, then v2>v1. Also, the min velocity needed is increasing.
